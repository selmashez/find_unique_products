```python
import pandas as pd
```

```python
# here I uploaded excel files and created dataframs
products1 = pd.read_excel("products_file_1_large.xlsx")
products2 = pd.read_excel("products_file_2_large.xlsx")

# Let's see dataframs
print(products1.head())
print(products2.head())
```

```python
# Concatenate the two DataFrames
Products = pd.concat([products1, products2], ignore_index=True)

Products.columns = ['all products']

print(Products)
```

```python
import transformers
transformers.logging.set_verbosity_error()
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

def find_unique_products(product_list, threshold=0.8):
    unique_products = []
    product_counts = {}

    for product in product_list:
        embedding = get_embedding(product)
        is_unique = True

        for unique_product in unique_products:
            unique_embedding = get_embedding(unique_product)
            similarity = cosine_similarity(embedding, unique_embedding)
            if similarity[0][0] > threshold:
                product_counts[unique_product] += 1
                is_unique = False
                break

        if is_unique:
            unique_products.append(product)
            product_counts[product] = 1

    return unique_products, product_counts


product_list = Products["all products"].tolist()


unique_products, product_counts = find_unique_products(product_list)

print(f"Total unique products: {len(unique_products)}")
for product, count in product_counts.items():
    print(f"Product: {product}, Count: {count}")
```
